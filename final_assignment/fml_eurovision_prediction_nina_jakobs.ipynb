# Can odds predict the position of the Eurovision Songcontest finals?

## Introduction

The annual Eurovision Songcontest ends with a final round of the 27 best contestants of the contest. Each year, bookmakers try to predict who is (not) going to win the contest. The lower the odds, the higher
the chance of winning the competition. The higher the odds, the less likely a contestant will win and the
more money you’ll get. The outcomes of the odds can be seen as a prediction of the final positions of the
competition. Different machine learning algorithms (e.g. Random Forest and K-Nearest Neighbor) have been used to discover if the bookmakers are
always right. For this case, we rely on the odds from 2015-2019 and 2021.

#### Odds from 2015-2019

```python
# Import libraries we need for webscraping and transformation to a json file
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import json
```

```python
# Create a driver for webscraping
DRIVER_PATH = 'C:/Users/ninaj/Desktop/chromedriver_win32 (1)/chromedriver.exe' # !! !This is a local variable
driver = webdriver.Chrome(executable_path=DRIVER_PATH)

driver.maximize_window()
driver.get('https://google.com') # Get the driver from Google

# Set up options
options = Options()
options.headless = True
options.add_argument("--window-size=1920,1200")
```

```python
# Define the years we're going to use. 2021 has been left out since 
#the URL for 2021 didn't inluce "-2021". Instead the year has been left empty. 
years = 2015, 2016, 2017, 2018, 2019
```

```python
# Create a new empty array. We're going to fill this with Eurovsion data. 
data = []
```

```python
# Create a for-loop to iterate through the years 2015-2019
for x in years:
    
    driver = webdriver.Chrome(options=options, executable_path=DRIVER_PATH)
    
    # The Eurovsion URL with a variable (x) for the years 2015-2019
    driver.get("https://eurovisionworld.com/odds/eurovision-" + str(x)) 

    # Define the table where we need the data from. 
    numbers = driver.find_elements_by_xpath("//tbody/tr")
    

    # Iterate through the number of contestants. Each row from the table is one contestant
    for row in numbers:

        # Loop through all positions
        position_data = ""
        for position in row.find_elements_by_css_selector("td.opo"):
            position_data = position.text
        
        # Loop through all contestants
        contestant_data = ""
        for contestant in row.find_elements_by_css_selector("td.odt.odtl"):
            contestant_data = contestant.text
        
        # Loop through all the odd data (contestant and position included, will be removed later)
        odds_data = [odds.text for odds in row.find_elements_by_css_selector("td")]
        
        # Create a dictionary
        esf_all = {
            "contestant": contestant_data,
            "position": position_data,
            "odds": odds_data
            } 

        # Append the dictionary with Eurovsion data to the empty data array
        data.append(esf_all)
   
        # Create the JSON file
        with open("esf_data_2015_2019.json", "w", encoding='utf-8' ) as outfile:
            json.dump(data, outfile, ensure_ascii=False)


driver.close()
```

## Predictive models

We need to split the data into a train and test set in order to train the model before creating a realistic prediction (test data). 

The prediction will be based on the Random Forest Regressor algorithm. Because the predicted value is not classified, but an actual value with variance, we're going to use the **Random Forest Regressor**. However, we're going to compare this with **K-Nearest Neighbor Regressor**. My assumption is that K-Nearest Neighbor will perform not as good as Random Forest. K-Nearest Neighbor will predict, as the name says, by looking at neighbors with the same factors/values. While Random Forest will try to find which odds predict the best, resulting in a better outcome. 

The Random Forest model finds the best X-variables (odds) for the best positions. In theory, you can also calculate which odd category (such as "UNIBET" and "BET365") are the best categories to earn money. The higher the values, the more money you win. 

In GitHub you can find the PDF of the poster I made for this case. Hopefully it helps you understanding the Random Forest model even better :-)
