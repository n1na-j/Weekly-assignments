### Retrieving Eurovision data by webscraping with Selenium

#### Eurovision data from 2015-2019

```python
# Import libraries we need for webscraping and transformation to a json file
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import json
```

```python
# Create a driver for webscraping
DRIVER_PATH = 'C:/Users/ninaj/Desktop/chromedriver_win32 (1)/chromedriver.exe' # !! !This is a local variable
driver = webdriver.Chrome(executable_path=DRIVER_PATH)

driver.maximize_window()
driver.get('https://google.com') # Get the driver from Google

# Set up options
options = Options()
options.headless = True
options.add_argument("--window-size=1920,1200")
```

```python
# Define the years we're going to use. 2021 has been left out since 
#the URL for 2021 didn't inluce "-2021". Instead the year has been left empty. 
years = 2015, 2016, 2017, 2018, 2019
```

```python
# Create a new empty array. We're going to fill this with Eurovsion data. 
data = []
```

```python
# Create a for-loop to iterate through the years 2015-2019
for x in years:
    
    driver = webdriver.Chrome(options=options, executable_path=DRIVER_PATH)
    
    # The Eurovsion URL with a variable (x) for the years 2015-2019
    driver.get("https://eurovisionworld.com/odds/eurovision-" + str(x)) 

    # Define the table where we need the data from. 
    numbers = driver.find_elements_by_xpath("//tbody/tr")
    

    # Iterate through the number of contestants. Each row from the table is one contestant
    for row in numbers:

        # Loop through all positions
        position_data = ""
        for position in row.find_elements_by_css_selector("td.opo"):
            position_data = position.text
        
        # Loop through all contestants
        contestant_data = ""
        for contestant in row.find_elements_by_css_selector("td.odt.odtl"):
            contestant_data = contestant.text
        
        # Loop through all the odd data (contestant and position included, will be removed later)
        odds_data = [odds.text for odds in row.find_elements_by_css_selector("td")]
        
        # Create a dictionary
        esf_all = {
            "contestant": contestant_data,
            "position": position_data,
            "odds": odds_data
            } 

        # Append the dictionary with Eurovsion data to the empty data array
        data.append(esf_all)
   
        # Create the JSON file
        with open("esf_data_2015_2019.json", "w", encoding='utf-8' ) as outfile:
            json.dump(data, outfile, ensure_ascii=False)


driver.close()
```

### Cleaning the Eurovision data

```python
# Import the libraries we need to create a dataframe from the JSON files
import pandas as pd
import json
```

```python
# Open the JSON file for the years 2015-2019
with open("esf_data_2015_2019.json", encoding='utf-8') as json_file:
    esf_data = json.load(json_file)
```

```python
# Create a df for 2015-2019
df_esf = pd.DataFrame(esf_data)

# Skip the first row, which doesn't include any useful data
df_esf = df_esf.drop([0, 1])
```

```python
# Create new columns per odd.We only need to do this once.
df_esf['BET365'] = df_esf['odds'].str[2]
df_esf['UNIBET'] = df_esf['odds'].str[3]
df_esf['YOU WIN'] = df_esf['odds'].str[4]
df_esf['BOYLE SPORTS'] = df_esf['odds'].str[5]
df_esf['CORAL'] = df_esf['odds'].str[6]
df_esf['SKY BET'] = df_esf['odds'].str[7]
df_esf['BET VICTOR'] = df_esf['odds'].str[8]
df_esf['TITAN BET'] = df_esf['odds'].str[9]
df_esf['PAF'] = df_esf['odds'].str[10]
df_esf['BETFAIR SPORT'] = df_esf['odds'].str[11]
df_esf['SPORTING BET'] = df_esf['odds'].str[12]
df_esf['INTERTOPS'] = df_esf['odds'].str[13]
df_esf['PADDY POWER'] = df_esf['odds'].str[14]
df_esf['BET FRED'] = df_esf['odds'].str[15]
df_esf['BETWAY'] = df_esf['odds'].str[16]
df_esf['STAN JAMES'] = df_esf['odds'].str[17]
df_esf['BWIN'] = df_esf['odds'].str[18]
df_esf['MATCHBOOK'] = df_esf['odds'].str[19]
df_esf['BETFAIR* EXCHANGE'] = df_esf['odds'].str[20]
df_esf['BETFAIR* EXCHANGE 2'] = df_esf['odds'].str[21] 
```

```python
# Remove the odd column. We don't need this anymore.
del df_esf['odds']
```

I noticed the bet category names differ each year. Since we wanted to focus on the values of the bets, the names don't really matter. However, we want to exclude values which may cause problems in the prediction process. Therefore, it seems logical to remove the BETFAIR EXCHANGE values. Because some years have more/less bet categories, the BETFAIR EXCHANGE column position has been used to retrieve the extra columns for the bets. For example: 2016 has an extra bet category, BETFAIR EXCHANGE is filled with the values of the extra bet category. The actual BETFAIR EXCHANGE values are removed. 

```python
# Create df for 2015. Only final positions are included
df_2015 = df_esf.iloc[0:27].reset_index()
df_2015['year'] = 2015
del df_2015['BETFAIR* EXCHANGE']
del df_2015['BETFAIR* EXCHANGE 2']

# Create df for 2016. Only final positions are included
df_2016 = df_esf.iloc[29:55].reset_index()
df_2016['year'] = 2016
del df_2016['BETFAIR* EXCHANGE 2']

# Create df for 2017. Only final positions are included
df_2017 = df_esf.iloc[57:83].reset_index()
df_2017['year'] = 2017
del df_2017['BET365']
del df_2017['BETFAIR* EXCHANGE']
del df_2017['BETFAIR* EXCHANGE 2']

# Create df for 2018. Only final positions are included
df_2018 = df_esf.iloc[102:128].reset_index()
df_2018['year'] = 2018
del df_2018['BET365']
del df_2018['UNIBET']
del df_2018['BETFAIR* EXCHANGE']
del df_2018['BETFAIR* EXCHANGE 2']

# Create df for 2019. Only final positions are included
df_2019 = df_esf.iloc[147:173].reset_index()
df_2019['year'] = 2019

# Add BETWAY for this year
del df_2019['BET365']
del df_2019['UNIBET']
# Remove the BET FAIR EXHANGES. These numbers have a larger range than the other odds, which can cause wrong assumptions later on. 
del df_2019['BETFAIR* EXCHANGE 2']
```
