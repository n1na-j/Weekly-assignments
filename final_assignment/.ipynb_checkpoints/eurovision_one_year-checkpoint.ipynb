### Eurovision Song Contest data for one year (2019)

```python
# Import needed libraries
import seaborn as sns #this is the plotting library I'll be using 
import pandas as pd #"as pd" means that we can use the abbreviation in commands
import matplotlib.pyplot as plt
import pandas as pd
import json
import tweepy
import requests
from bs4 import BeautifulSoup
```

```python
df = pd.read_csv('contestants.csv')
df.head()
```

```python
# Filter by year
df_2019 = (df[df["year"] == 2019])

# Show only the columns which are necessary for the prediction
df_2019 = df_2019[["year", "to_country", "performer", "song", "place_final"]]


# Sort final results
df_2019.sort_values(by=['place_final'])
df_2019.head()
```

```python
# Bookmakers 

# URL we're going to scrape
url = "https://eurovisionworld.com/odds/eurovision-2019"

# Accept cookies
s = requests.Session()
s.cookies.update({'__hs_opt_out': 'no'})
s.get(url)  # Automatically uses the session cookies

# Iterate through years

#Load page
page = requests.get(url)

soup = BeautifulSoup(page.content, "html.parser")

# Get the table
table = soup.find("div").get_text()
print(table)

# data = []


```
