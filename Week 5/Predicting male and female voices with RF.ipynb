# Using the Random Forest algorithm to predict if a voice is from a woman or man
For this assignment, I'm using the Random Forest algorithm to predict if a voice is recognized as a female or a male voice. According to the [documentation](https://www.kaggle.com/primaryobjects/voicegender/home), Random Forest should be able to give a 100% / 98% accuracy whereas the baseline (always predict male) is 50% / 50%

 

## Importing the data

```python
# Import the neccessary libraries
import seaborn as sns
import sklearn as sk
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split #We need this to split the data
from sklearn.preprocessing import normalize # We need this to normalize our data
from sklearn.ensemble import RandomForestClassifier # We need this to use the RF classifier
from sklearn.metrics import confusion_matrix # We need this to make a confusion matrix
from sklearn.metrics import classification_report # We need this to calculate the accuracy, precision and recall
```

```python
df = pd.read_csv('voice.csv')
df.head()
```

We don't need to use all the variables, only the ones we find interesting. Based on general knowledge, we can say that we definitely need the following X variables:

- meanfreq: mean frequency (in kHz)
- IQR: interquantile range (in kHz)
- meandom: average of dominant frequency measured across acoustic signal
- meanfun: average of fundamental frequency measured across acoustic signal
- sp.ent: spectral entropy
- dfrange: range of dominant frequency measured across acoustic signal

And for y we use:
- label: male or female


Let's check how the variables look like in a violin plot:

```python
# Create violin plots since we focus on the categorical y variable label
sns.set(style="whitegrid")

sns.violinplot(x='label', y='meanfreq', data=df)
plt.title("Relation between gender and meanfreq")
plt.ylabel("meanfreq")
plt.xlabel("Label")
plt.show()

sns.violinplot(x='label', y='IQR', data=df)
plt.title("Relation between gender and IQR")
plt.ylabel("IQR")
plt.xlabel("Label")
plt.show()


sns.violinplot(x='label', y='meandom', data=df)
plt.title("Relation between gender and meandom")
plt.ylabel("meandom")
plt.xlabel("Label")
plt.show()

sns.violinplot(x='label', y='meanfun', data=df)
plt.title("Relation between gender and meanfun")
plt.ylabel("meanfun)")
plt.xlabel("Label")
plt.show()

sns.violinplot(x='label', y='sp.ent', data=df)
plt.title("Relation between gender and sp.ent")
plt.ylabel("sp.ent")
plt.xlabel("Label")
plt.show()

sns.violinplot(x='label', y='dfrange', data=df)
plt.title("Relation between gender and dfrange")
plt.ylabel("modindx")
plt.xlabel("Label")
plt.show()
```

You can see the differences in male and female voices clearly based on the violin plots.

We expect to find 50% men and 50% women in the column 'label'. To validate this, we can count the amount of men and women from this column. 

```python
df['label'].value_counts()
```

Gender is indeed equally divided. 

Let's split the dataset into a training and test set

```python
from sklearn.preprocessing import normalize #get the function needed to normalize our data.

X = df[['meanfreq', 'IQR', 'meandom', 'meanfun', 'sp.ent', 'dfrange']] #create the X matrix
y = df['label'] #create the y-variable
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1) #split the data, store it into different variables
X_train.head() # Show the head of the trainingset
```

## Training the algorithm

Now we can train the RF (Random Forest) algorithm. The random_state sets the randomness for the RF algorithm so we can predict on stable, randomly sorted data. 

The number of trees (n_estimators) is set to 100, which is the default number of trees we'll be using later on for the sklearn library. 

First, we're going to use the train dataset in order to train the algorithm. 
We train the data on the selected variables from the dataset to predict if a voice is from a male or female.  

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(random_state=1, n_estimators=100) #RF is a random algorithm, so to get the same results we need to use random_state
rf = rf.fit(X_train, y_train)
```

## Evaluating the model

Before we use the test dataset, we're going to evaluate the model by a confusion matrix. With the confusion matrix we can calulcate the accuracy, precision and recall. 

```python
from sklearn.metrics import confusion_matrix
y_pred = rf.predict(X_test) #the predicted values
cm = confusion_matrix(y_test, y_pred) #creates a "confusion matrix"
cm = pd.DataFrame(cm, index=['male (actual)', 'female (actual)'], columns = ['male (pred)', 'female (pred)']) #label and make df
cm
```

Since the labels for male/female are equally divided, I expected to see less difference between male/female. However, the documentation also shows a higher accuracy for men than women. 

Let's look at the calculations for the accuracy, precision and recall. 

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))

# F1 score is the precision and recall mean
# Recall is very important here in this case bc the fraud can next month be different
# Precision is very important for the corona tests (% of correct positives)
# What is acceptable depends on the case!
```

None of the values is below 0.97 which seems quite close to the expected outcomes. The accuracy is 0.98 which is also close to the RF outcome from the documentation (100% accuracy for men and 98% for women). The f1-score is the mean of the precision and recall, which is equal for both men and women. 

We can find small differences between the two labels. The precision is 0.01 lower for women than for men and the recall for men is 0.01 lower than for women. 

Overall, the algorithm scores really close to the expected outcomes.
